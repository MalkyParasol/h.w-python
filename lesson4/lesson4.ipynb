תרגיל 2: גרף עמודות

מטרה: להציג את כמות המכירות לפי קטגוריה.
נתונים: מכירות של 4 קטגוריות מוצרים בחנות.

```python
categories = ['shoes','dress','pizza','coffee']
values = [2,45,3,7]
```

```python
plt.bar(categories, values)
plt.title('my shop')
plt.xlabel('Categories')
plt.ylabel('Values')
```

תרגיל 3: גרף פיזור

    מטרה: לחקור את הקשר בין זמן לימוד לציונים במבחן.
    נתונים: שעות לימוד וציונים של תלמידים.

```python
import numpy as np
```

```python
time = [2,4,6,7,2,9,5,7,8,4,3]
marks=[100,95,85,69,48,68,97,99,100,12,56]
colors = np.random.rand(11) # 50 צבעים אקראיים
sizes = 1000 * np.random.rand(11) # 50 גדלים אקראיים לנקודות
```

```python
plt.scatter(time,marks, c=colors, s=sizes, alpha=0.5, cmap='viridis')
plt.colorbar()  # הצגת סרגל הצבעים
plt.title('Scatter Plot Example')
plt.xlabel('time axis')
plt.ylabel('marks axis')
```

תרגיל 4: היסטוגרמה

    מטרה: לנתח את התפלגות גילאים בקבוצת נבדקים.
    נתונים: גילאים של אנשים בקבוצת מחקר.


```python
ages = [3,65,23,8,32,5,8,34,34,32,12,43,86]
plt.hist(ages, bins=[20, 30, 40, 50, 60, 70], edgecolor='black')

plt.title('Age Distribution')
plt.xlabel('Ages')
plt.ylabel('Number of People')
```

תרגיל 5: גרפים מרובים בחלון אחד (Subplots)

    מטרה: להשוות בין צריכת אנרגיה לפי שנים ומקורות.
    נתונים: צריכת אנרגיה לפי שנים ומקורות אנרגיה שונים.

```python
years = range(2010, 2021)
consumption_oil = [100, 105, 110, 120, 125, 130, 135, 140, 145, 150, 155]
consumption_gas = [80, 82, 85, 88, 90, 95, 100, 105, 110, 115, 120]
consumption_renewable = [20, 25, 30, 35, 40, 45, 50, 60, 70, 80, 90]
```

```python
plt.plot(years, consumption_oil, '-b', label='consumption_oil')
plt.plot(years, consumption_gas, ':r', label='consumption_gas')
plt.plot(years, consumption_renewable, ':a', label='consumption_renewable')
```
